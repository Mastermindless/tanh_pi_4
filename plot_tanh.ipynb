```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Define the explicit function
def symetric_tanh_pi_4(x):
    return ((np.exp(x)) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# Generate x values and compute y values for the function
x_values = np.linspace(-10, 10, 1000)
y_values_explicit = symetric_tanh_pi_4(x_values)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values_explicit, label=r"Conventional $\tanh_(x)$", color="blue")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Conventional $\ttanh_(x)$", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel(r"y", fontsize=12)
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```
